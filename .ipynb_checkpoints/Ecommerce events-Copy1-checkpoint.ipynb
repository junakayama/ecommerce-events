# 4. When is the site most busy?

```python
time = df.date.apply(lambda x: x.time())
```

```python
time.value_counts().plot()
```

```python
nu_weekday = df.date.apply(lambda x: x.weekday())
```

```python
nu_weekday.head()
```

```python
weekdays = []
for n in nu_weekday:
    if n==0:
        weekday = 'monday'
    if n==1:
        weekday = 'tuesday'
    if n==2:
        weekday = 'wednesday'
    if n==3:
        weekday = 'thursday'
    if n==4:
        weekday = 'friday'
    if n==5:
        weekday = 'saturday'
    if n==6:
        weekday = 'sunday'
    weekdays.append(weekday)
```

```python
df['weekday'] = weekdays
```

```python
df.weekday.value_counts().plot(kind='bar')
```
